```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

# Loading the Data set

```python
def load_data(file):
    try : 
        data = pd.read_csv(file,sep="\t")
        print(f"{file} Data Loaded Successfully!")
        print("\n Data head : ")
        display(data.head())
        print("\n Basic Information :")
        data.info() 
        print("\n Summary Statistics :")
        display(data.describe())
        print("\n Skewness :")
        print(data.skew(numeric_only = True))
        print("\n Kurtosis :")
        print(data.kurt(numeric_only = True))
        return data
        
    except Exception as e :
        print(f"Error loading the dataset : {e}")
    return None
    
MarketCampaign = load_data("marketing_campaign_before_clean.csv")
```

# Data Cleanig And Preprocessing

```python
def clean_data(data, remove_outliers=False):
    # Duplicate Value Treatment
    print(f"Total no of rows: {len(data)}")
    print(f"Duplicate Values: {data.duplicated().sum()}")
    data.drop_duplicates(inplace=True)
    print(f"Duplicates removed. Rows Remaining: {len(data)}\n")
    
    # Missing value treatment
    print("Missing Values (%):")
    print(round(data.isnull().sum() / len(data) * 100, 2), "\n")
    data.dropna(inplace=True)
    print("After Removing Null Values:")
    print(round(data.isnull().sum() / len(data) * 100, 2), "\n")
    
    # Outlier detection and treatment
    numeric_data = data.select_dtypes(include=np.number)
    outlier_indices = set()
    
    for col in numeric_data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        LW = Q1 - 1.5 * IQR
        UW = Q3 + 1.5 * IQR
        outliers = (data[col] < LW) | (data[col] > UW)
        out_pct = outliers.sum() / len(data) * 100

        if outliers.sum() > 0:
            print(f"{col}: {out_pct:.2f}% outliers")
            if remove_outliers:
                outlier_indices.update(data[outliers].index)
        else:
            print(f"{col}: No outliers")
    
    if remove_outliers and outlier_indices:
        print(f"\nRemoving {len(outlier_indices)} rows with outliers.")
        data.drop(index=outlier_indices, inplace=True)
        print(f"Rows remaining after outlier removal: {len(data)}\n")
    
    return data
```

```python
cleaned_data = clean_data(MarketCampaign, remove_outliers=True)
```

```python
df = cleaned_data
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)
df["Customer_Tenure_Days"] = (datetime.now() - df["Dt_Customer"]).dt.days
```

```python
df["Total_Spending"] = df[["MntWines", "MntFruits", "MntMeatProducts","MntFishProducts", "MntSweetProducts", "MntGoldProds"]].sum(axis=1)
```

# Exploratory Data Analysis

```python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Count plot of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 1. Age Distribution
df["Age"] = 2025 - df["Year_Birth"]
sns.histplot(df["Age"], bins=20)
plt.title("Customer Age Distribution")
plt.show()
```

```python
# 2. Response Rate by Education
sns.barplot(x="Education", y="Response", data=df)
plt.title("Response by Education Level")
plt.show()
```

```python
# 3. Spending Patterns by Marital Status
sns.boxplot(x="Marital_Status", y="Total_Spending", data=df)
plt.title("Spending by Marital Status")
plt.show()
```

```python
# Relationship between Income and Spending
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Income', y='Total_Spending', hue='Response')
plt.title('Income vs Total Spend by Campaign Response')
plt.show()
```

```python
# Response Rate by Marital Status / Education
plt.figure(figsize=(12, 5))
sns.barplot(data=df, x='Marital_Status', y='Response')
plt.title('Response Rate by Marital Status')
plt.show()

plt.figure(figsize=(12, 5))
sns.barplot(data=df, x='Education', y='Response')
plt.title('Response Rate by Education')
plt.show()
```
