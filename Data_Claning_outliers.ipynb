```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings                                           
warnings.filterwarnings('ignore')
```

# Loading the Data set

```python
def load_data(file):
    try : 
        data = pd.read_csv(file,sep="\t")
        print(f"{file} Data Loaded Successfully!")
        print("\n Data head : ")
        display(data.head())
        print("\n Basic Information :")
        data.info() 
        print("\n Summary Statistics :")
        display(data.describe())
        print("\n Skewness :")
        print(data.skew(numeric_only = True))
        print("\n Kurtosis :")
        print(data.kurt(numeric_only = True))
        return data
        
    except Exception as e :
        print(f"Error loading the dataset : {e}")
    return None
    
MarketCampaign = load_data("marketing_campaign_before_clean.csv")
```

# Data Cleanig And Preprocessing

```python
def clean_data(data):
    # Duplicate Value Treatment
    print(f"Total no of rows: {len(data)}")
    print(f"Duplicate Values: {data.duplicated().sum()}")
    data.drop_duplicates(inplace=True)
    print(f"Duplicates removed. Rows Remaining: {len(data)}\n")
    
    # Missing value treatment
    print("Missing Values (%):")
    print(round(data.isnull().sum() / len(data) * 100, 2), "\n")
    data.dropna(inplace=True)
    print("After Removing Null Values:")
    print(round(data.isnull().sum() / len(data) * 100, 2), "\n")
    return data
```

```python
cleaned_data = clean_data(MarketCampaign)
```

```python
# To Remove Outliers
cleaned_data = cleaned_data[cleaned_data["Income"] < 200000]
```

```python
# Replace some of the values in "Martial Status"
cleaned_data['Marital_Status'].replace({
    'Married': 'Together',
    'Alone': 'Single',
    'Divorced': 'Single',
    'Widow': 'Single',
    'Absurd': 'Other',
    'YOLO': 'Other'
}, inplace=True)

# Review after replace
cleaned_data['Marital_Status'].value_counts()
```

```python
# Replace "Education" values to be more useful
cleaned_data['Education'].replace({
    'Graduation': 'Postgraduate',
    'PhD': 'Postgraduate',
    'Master': 'Postgraduate',
    '2n Cycle': 'Undergraduate',
    'Basic': 'Undergraduate'
}, inplace=True)

# Review after replace
cleaned_data['Education'].value_counts()
```

```python
df = cleaned_data
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)
df["Customer_Tenure_Days"] = (datetime.now() - df["Dt_Customer"]).dt.days
```

```python
df["Total_Spending"] = df[["MntWines", "MntFruits", "MntMeatProducts","MntFishProducts", "MntSweetProducts", "MntGoldProds"]].sum(axis=1)
```

# Exploratory Data Analysis

```python
numeric_cols = df.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Count plot of {col}')
    plt.xticks(rotation=45)
    plt.show()
```
